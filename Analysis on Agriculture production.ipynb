```python
grouped_df = df.groupby("SubCounty").sum() #we are groupping the Subcounty column to find the sum
print(grouped_df.head())
```

```python
# performing aggregation where we are assigning variable agg_df to more than one object
agg_df = df.groupby('SubCounty').agg({
    'Farming_Activities': 'sum',
    'Crop_Production': 'sum',
    'Livestock_Production': 'sum',
    'Aquaculture': 'sum',
    'Fishing': 'sum',
    'Irrigation': 'sum'
})
print(agg_df.head())
```

```python
agg_df
```

```python
grouped_df
```

```python
# Adding a new column for total agricultural activities
'''
In Python, aggregation is a design principle that allows one object to contain or reference another object.
It's a form of object composition where an object is composed of one or more other objects,
which it uses as its parts. Aggregation represents a "has-a" relationship, 
meaning that an object "has a" reference to another object. 
'''
df['Total_Agriculture'] = (df['Farming_Activities'] +
                           df['Crop_Production'] +
                           df['Livestock_Production'] +
                           df['Aquaculture'] +
                           df['Fishing'] +
                           df['Irrigation'])
print(df.head())
```
